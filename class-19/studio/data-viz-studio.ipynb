```python
# Use your Python skills to create a visualization here.
# To understand the relationship between average rating and ratings count
# Creating a dataframe with 'average_rating' and 'ratings_count' columns
ratings_data = df[['average_rating', 'ratings_count']]

# Creating a scatterplot to show how the average rating of a book relates to the number of ratings it has received.
plt.scatter(x=ratings_data['average_rating'], y=ratings_data['ratings_count'])
plt.xlabel('average_rating')
plt.ylabel('ratings_count')
plt.show()

# The scatterplot below shows there is no direct relationship between ratings count and average rating of books. Books with high ratings count
# don't necessarily have a high or low rating. Most of the books have average ratings concentrated between 3.5 and 4.5.
```
